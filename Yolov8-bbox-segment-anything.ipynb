```python
bbox = {
           "x": "263.31",
            "y": "763.03",
            "width": "144.80",
            "height": "58.95"
        }

x, y = float(bbox['x']), float(bbox['y'])
hx, hy = float(bbox['width']), float(bbox['height']), 
input = [x - hx/2, y - hy/2, x + hx/2, y + hy/2]
print(input)

result = SegManualMaskPredictor().save_image(
    source="test.jpg",
    model_type="vit_l",
    input_point=None,
    input_label=None,
    input_box= [input],
    multimask_output=False,
)

from IPython.display import Image
Image(result)
```
